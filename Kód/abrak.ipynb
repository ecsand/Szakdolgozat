```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
#éves kalóztámadások
inc=pd.read_csv("C:\\Users\\Andris\\Documents\\Szakdoga\\scaled\\incm.csv")
inc.head()
```

```python

inc.index=pd.to_datetime(inc['Unnamed: 0'])
incc=inc['incidents_per_month']
incc
```

```python
inc2=pd.DataFrame()
inc2.index=inc.index
inc2["inc"]=incc
inc3=inc2.groupby(inc.index.year)['inc'].sum()
inc3
```

```python
geo=pd.read_excel("C:\\Users\\Andris\\Documents\\Szakdoga\\szószok\\api_ideig.xlsx")
late=pd.read_excel("C:\\Users\\Andris\\Documents\\Szakdoga\\df\\dfveg_latest_2.xlsx")

geo.drop(columns='Unnamed: 0', inplace=True)
late.drop(columns=['Unnamed: 0', 0, 2, 4, 5, 7, 8], inplace=True)
incidents=pd.concat([late, geo], axis=1)
incidents.head()
df=incidents
df1 = df[~df[9].str.contains(r"(?=.*Malacca)(?=.*Singapore)|SOMS|S0MS", na=False)]
df1.index=pd.to_datetime(df1.otket)

dff=df1.groupby(df1.index.year)['lat1'].size()
dff


df2 = df[df[9].str.contains(r"(?=.*Malacca)(?=.*Singapore)|SOMS|S0MS", na=False)]
df2.index=pd.to_datetime(df2.otket)

dff2=df2.groupby(df2.index.year)['lat1'].size()
```

```python
plt.plot(dff2.index, dff2)

plt.title('A SOMS-beli kalóztámadások/ARAS-incidensek éves alakulása (db/év)')
plt.legend()
plt.show()
```

```python
plt.plot(dff2.index, dff2, label="SOMS")
plt.plot(dff.index, dff, label="SOMS-t leszámítva")
plt.title('Az ázsiai kalóztámadások/ARAS-incidensek éves alakulása (db/év)')
plt.legend()
plt.show()
```

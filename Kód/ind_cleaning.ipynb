```python
q=grdp.columns.tolist()

q2=[]
for x in range(2, len(q)):
    if q[x]!=q[x-1]:
        q2.append("Q1")
        i=2
    else:
        if i!=5:

            q2.append("Q"+str(i))
        else:
            q2.append("Annual")
        i=i+1
q1=['Province', 'Industrial Origin']

qq=q1+q2
print(qq)
```

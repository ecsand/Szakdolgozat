```python
gdp_qtr=pd.read_csv("C:\\Users\\Andris\\Documents\\Szakdoga\\szószok\\Malaysia\\gdp_qtr_nominal.csv")
gdp_qtr.index=gdp_qtr['date']
gdp_qtr.drop(columns='date', inplace=True)
gdp_qtr_abs=gdp_qtr.loc[gdp_qtr["series"]=="abs"]
gdp_qtr_growth=gdp_qtr.loc[gdp_qtr["series"]=="growth_qoq"]
gdp_qtr_abs.drop(columns='series', inplace=True)
gdp_qtr_growth.drop(columns='series', inplace=True)
gdp_qtr_abs.columns=["nominal gdp"]
gdp_qtr_growth.columns=["nominal gdp growth"]
def inter(df):

    df.index=pd.to_datetime(df.index)

    mon = pd.date_range(start=df.index.min(), end=pd.Timestamp('2024-05-01'), freq='M')
    df2= pd.DataFrame(index=mon, columns=df.columns)
    for x in df.index.year:
        for y in df.index.month:
            df2.loc[(df2.index.year == x) & (df2.index.month == y)] = df.loc[(df.index.year == x) & (df.index.month == y)].values


    #interpolation
    df2 = df2.apply(pd.to_numeric, errors='coerce')
    df2_time = df2.interpolate(method='time')
    df2_lin=df2.interpolate(method='linear')
    l=df2_lin.columns
    t=df2_time.columns
    l1=[]
    t1=[]
    for x, y in zip(l, t):
        l1.append(x+ " (linear)")
        t1.append(y+ " (time)")
    df2_lin.columns=l1
    df2_time.columns=t1
    df2_fin=pd.concat([df2_lin, df2_time], axis=1)

    return df2_fin


gdp_qtr_growth=inter(gdp_qtr_growth)
gdp_qtr_abs=inter(gdp_qtr_abs)
gdp_qtr_abs.tail()
```

```python
kl=pd.concat([klasszres, klasszres_thai], axis=0)
kl.drop(columns="Unnamed: 0", inplace=True)

t2=[]
tr2=[]
for x, y in zip(kl['test acc'], kl["training acc"]):
    if x>1:
        t2.append(x/100)
    else:
        t2.append(x)

    if y>1:
        tr2.append(y/100)
    else:
        tr2.append(y)
kl.drop(columns=['test acc', 'training acc'], inplace=True)
kl["test"]=t2
kl["training"]=tr2
kl.iloc[:40]
```

```python
kl=kl.sort_values(by="test", ascending=False)
kl=kl[kl["training"]>=kl['test']]
```

```python
kl=kl[kl["training"]!=1]
```

```python
kat=[]
for x in kl.nev:
    if "gaussian" in x:
        kat.append("Naive bayes")
    elif "tree" in x:
        kat.append("Decision tree")
    elif "boosting" in x:
        kat.append("Gradient boosting classifier")
    elif "svm" in x:
        kat.append("SVM")
    elif "ridge" in x:
        kat.append("Ridge")
    elif "logreg" in x:
        kat.append("Logistic regression")

kl["kat"]=kat
```

```python
k=kl.sort_values(by="test", ascending=False)
k.iloc[0:30]
```

```python
kl.drop(columns="nev", inplace=True)
```

```python
a=kl.groupby("kat").mean()
```

```python
gdp_qtr=pd.read_excel("C:\\Users\\Andris\\Documents\\Szakdoga\\szószok\\Singapore (idk, kell-e)\\M015651.xlsx")
df=gdp_qtr
df=df.iloc[9:, :]
df.columns=df.iloc[0]
df=df.iloc[1:30, :]
df=df.T
df.columns=df.iloc[0]
df=df.iloc[1:, :]
i=[]
for x in df.index:
    s=str(x).split(" ")
    a=s[0]
    b=s[1]
    if b=="1Q":
        i.append(a+"-03-01")
    elif b=="2Q":
        i.append(a+"-06-01")
    elif b=="3Q":
        i.append(a+"-09-01")
    elif b=="4Q":
        i.append(a+"-12-01")
    else:
        i.append(str(x))


df.index=pd.to_datetime(i)
df=df.iloc[::-1]

mon = pd.date_range(start=df.index.min(), end=pd.Timestamp('2024-07-01'), freq='M')
df2= pd.DataFrame(index=mon, columns=df.columns)
for x in df.index.year:
    for y in df.index.month:
        df2.loc[(df2.index.year == x) & (df2.index.month == y)] = df.loc[(df.index.year == x) & (df.index.month == y)].values

df2 = df2.apply(pd.to_numeric, errors='coerce')
df2_time = df2.interpolate(method='time')
df2_lin=df2.interpolate(method='linear')
l=df2_lin.columns
t=df2_time.columns
l1=[]
t1=[]
for x, y in zip(l, t):
    l1.append(x+ " (linear)")
    t1.append(y+ " (time)")
df2_lin.columns=l1
df2_time.columns=t1
df2_fin=pd.concat([df2_lin, df2_time], axis=1)
gdp_qtr=df2_fin
```

```python
unc=pd.read_excel("C:\\Users\\Andris\\Documents\\Szakdoga\\szószok\\Singapore (idk, kell-e)\\WUISGP.xls")
unc=unc.iloc[9:, :]
unc.columns=unc.iloc[0]
unc=unc.iloc[1:, :]
unc
def inter2(df):
 

    df.index=pd.to_datetime(df["observation_date"])

    mon = pd.date_range(start=df.index.min(), end=pd.Timestamp('2024-08-01'), freq='M')
    df2= pd.DataFrame(index=mon, columns=df.columns)
    for x in df.index.year:
        for y in df.index.month:
            df2.loc[(df2.index.year == x) & (df2.index.month == y)] = df.loc[(df.index.year == x) & (df.index.month == y)].values

    df2.drop(columns=["observation_date"], inplace=True)
    print(df2.tail())
    #interpolation
    df2 = df2.apply(pd.to_numeric, errors='coerce')
    df2_time = df2.interpolate(method='time')
    df2_lin=df2.interpolate(method='linear')
    l=df2_lin.columns
    t=df2_time.columns
    l1=[]
    t1=[]
    for x, y in zip(l, t):
        l1.append("uncertainty (linear)")
        t1.append("uncertainty (time)")
    df2_lin.columns=l1
    df2_time.columns=t1
    df2_fin=pd.concat([df2_lin, df2_time], axis=1)

    return df2_fin


dfuncertainty=inter2(unc)
dfuncertainty.tail()
```

```python
import pandas as pd
import os
from openpyxl import load_workbook
import numpy as np
import re
```

```python
dfwb=pd.read_excel("C:\\Users\\Andris\\Documents\\Szakdoga\\sz√≥szok\\Thailand\\P_Data_Extract_From_World_Development_Indicators.xlsx") 


dfwb=dfwb.T

dfwb.columns=dfwb.iloc[2]
dfwb=dfwb.iloc[5:, :]
i= []
for x in dfwb.index:
    i.append(str(x).split(" [")[0])

dfwb.index = pd.to_datetime(i, format='%Y')
mon = pd.date_range(start=pd.Timestamp('2008-12-01'), end=pd.Timestamp('2024-01-01'), freq='M')
dfwbmon = pd.DataFrame(index=mon, columns=dfwb.columns)
dfwbmon.tail()

for x in dfwb.index.year:

    dfwbmon.loc[(dfwbmon.index.year == x) & (dfwbmon.index.month == 12)] = dfwb.loc[dfwb.index.year == x].values

dfwbmon.tail()

df=dfwbmon

df = df.apply(pd.to_numeric, errors='coerce')
dfint_time = df.interpolate(method='time')
dfint_lin=df.interpolate(method='linear')
dfwb=pd.concat([dfint_lin, dfint_time], axis=1)
```

```python
def one(df):
    df.index = df.index.map(lambda x: x.replace(day=1))
    return df

thai=one(dfwb)
thai
```

```python
for x in thai.columns:
    if str(x)=="NaN":
        thai.drop(columns=x, inplace=True)
thai
```

```python
df=thai
df=df.iloc[1:]
c2=[]
for x in df.columns:
    c2.append("Thailand: "+x)
df.columns=c2


```
